```python
df = pd.DataFrame({'company':['biodiscovery','TIGR','SAIC(NCI)','MSC(NIAID)','CSRA(NCI)'],'years':[1.5,7.5,2.5,1.1,2.1]})
df = df.set_index('company')
df.plot(kind='bar',color='green')
plt.title('work places and years')
plt.text(2,4,'total years: 15',fontsize=16)
```

```python
df.years = df.years*10
df.years = df.years.astype(int)
df.years.plot(kind='pie')
plt.axis('equal')
plt.title('work years distribution')
```
